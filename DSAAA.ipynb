```python
# Classic example: Factorial
def factorial(n):
    # Base case - stops recursion
    if n == 0 or n == 1:
        return 1
    
    # Recursive case - calls itself
    return n * factorial(n - 1)

res=factorial(5)
print(res)
```

```python
def two_sum(arr, target):
    # Step 1: Setup two pointers
    left = 0
    right = len(arr) - 1
    
    # Step 2: Loop until pointers meet
    while left < right:
        # Calculate current sum
        current_sum = arr[left] + arr[right]
        
        # Check against target
        if current_sum == target:
            return [left, right]  # Found!
        elif current_sum < target:
            left += 1   # Need bigger sum
        else:  # current_sum > target
            right -= 1  # Need smaller sum
    
    # Step 3: Not found
    return [-1, -1]
# Test
arr = [1, 2, 4, 5, 6]
target = 7
result = two_sum(arr, target)
print(f"Indices: {result}")  # Output: [2, 3]
print(f"Values: {arr[result[0]]}, {arr[result[1]]}")  # Output: 4, 5

arr = [1, 2, 4, 5, 6]
target = 9
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def has_cycle(head):
    if not head:
        return False
    
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True
    
    return False

node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node2  # Creates cycle!

print(has_cycle(node1))  # Output: True

node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)

node1.next = node2
node2.next = node3
node3.next = None

print(has_cycle(node1))  # Output: False
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeNthFromEnd(head, n):
    # Create dummy node
    dummy = ListNode(0)
    dummy.next = head
    
    first = dummy
    second = dummy
    
    # Create gap of n+1
    for i in range(n + 1):
        second = second.next
    
    # Move both together
    while second:
        first = first.next
        second = second.next
    
    # Remove node
    first.next = first.next.next
    
    return dummy.next

# Helper function to create linked list
def create_list(arr):
    dummy = ListNode(0)
    current = dummy
    for i in arr:
        current.next = ListNode(i)
        current = current.next
    return dummy.next

# Helper function to print linked list
def print_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    print(" → ".join(map(str, result)))


head = create_list([1, 2, 3, 4, 5])
print("Before:", end=" ")
print_list(head)  

result = removeNthFromEnd(head, 2)
print("After:", end=" ")
print_list(result)  
```

```python
def backspaceCompare(s: str, t: str):
    i, j = len(s) - 1, len(t) - 1
    skip_s = skip_t = 0
    
    while i >= 0 or j >= 0:
        # Find next valid char in s
        while i >= 0:
            if s[i] == '#':
                skip_s += 1
                i -= 1
            elif skip_s > 0:
                skip_s -= 1
                i -= 1
            else:
                break
        
        # Find next valid char in t
        while j >= 0:
            if t[j] == '#':
                skip_t += 1
                j -= 1
            elif skip_t > 0:
                skip_t -= 1
                j -= 1
            else:
                break
        
        # Compare characters
        if i >= 0 and j >= 0:
            if s[i] != t[j]:
                return False
        elif i >= 0 or j >= 0:
            return False  # One finished, other didn't
        
        i -= 1
        j -= 1
    
    return True

print(backspaceCompare("ab#c", "ad#c")) 
```

```python
#STACK

class Stack:
    def __init__(self):
        self.items=[]

    def push(self,item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        raise IndexError("Stack is empty")

    def is_empty(self):
        return len(self.items)==0

    def size(self):
        return len(self.items)

stack1=Stack()
stack2=Stack()
stack1.push(1)
stack1.push(2)
stack1.push(3)
stack1.pop()
stack2.push(10)
print(stack1.items)
print(stack2.items)
```
